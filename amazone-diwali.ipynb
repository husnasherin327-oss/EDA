## outlier detection

```python
for i in (df.select_dtypes(['int','float'])):
    sns.boxplot(x=i,data=df)
    plt.show()
```

#### IQR

```python
outliers = []

for col in df.select_dtypes(include=['int','float']):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr

    # values which are out of range
    o = df[(df[col] < lower) | (df[col] > upper)][col]

    print(f"Column: {col}")
    print(f"Outliers: {o.tolist()}\n")

    outliers.extend(o.tolist())

```
